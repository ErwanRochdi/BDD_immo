```python
import pandas as pd
import numpy as np
```

```python
df_donnees = pd.read_excel('Les_donnees.xlsx')
```

```python
df_donnees_wo_nan = df_donnees.dropna(axis=1, how="all", thresh=None, subset=None, inplace=False)
df_donnees
```

```python
df_mutation = df_donnees[['Date mutation', 'Nature mutation']].drop_duplicates()
df_mutation = df_mutation.reset_index(drop=True)
df_mutation['idMut'] = df_mutation.index.astype(str)+'_M'
df_mutation
```

```python
df_type = df_donnees[['Type local', 'Code type local']].drop_duplicates()
df_type = df_type.reset_index(drop=True)
df_type['idType'] = df_type.index.astype(str)+'_T'
df_type
```

```python
df_voie = df_donnees[['No voie', 'B/T/Q', 'Type de voie', 'Voie']].drop_duplicates()
df_voie = df_voie.reset_index(drop=True)
df_voie['idVoie'] = df_voie.index.astype(str)+'_V'
df_voie
```

```python
df_commune = df_donnees[['Commune', 'Code departement']].drop_duplicates()
df_commune = df_commune.reset_index(drop=True)
df_commune['idCommune'] = df_commune.index.astype(str)+'_C'
df_commune
```

```python
df_donnees = df_donnees.merge(df_voie, on = ['No voie', 'B/T/Q', 'Type de voie', 'Voie'], how = 'outer')
df_donnees = df_donnees.merge(df_commune, on = ['Commune', 'Code departement'], how = 'outer')
df_donnees = df_donnees.merge(df_mutation, on = ['Date mutation', 'Nature mutation'], how = 'outer')
df_donnees = df_donnees.merge(df_type, on = ['Type local', 'Code type local'], how = 'outer')
df_donnees
```

```python
df_donnees.columns
```

```python
df_donnees = df_donnees.drop(columns = ['Code service CH', 'Reference document', '1 Articles CGI',
       '2 Articles CGI', '3 Articles CGI', '4 Articles CGI', '5 Articles CGI',
       'No disposition', 'Code type de voie', 'Code voie', 'Code ID commune', 'Code postal', 'Code commune', 'Prefixe de section', 'Section', 'No plan', 'No Volume',
       '1er lot', 'Surface Carrez du 1er lot', '2eme lot',
       'Surface Carrez du 2eme lot', '3eme lot', 'Surface Carrez du 3eme lot',
       '4eme lot', 'Surface Carrez du 4eme lot', '5eme lot',
       'Surface Carrez du 5eme lot', 'Nombre de lots', 'Identifiant local', 'Nature culture',
       'Nature culture speciale', 'Surface terrain'])
```

```python
df_adresse = df_donnees[['idCommune', 'idVoie']].drop_duplicates()
df_adresse = df_adresse.reset_index(drop=True)
df_adresse['idAdresse'] = df_adresse.index.astype(str)+'_A'
df_adresse
```

```python
df_donnees_for_dflocal = df_donnees.merge(df_adresse, on = ['idVoie', 'idCommune'], how = 'outer')
df_donnees_for_dflocal
```

```python
df_local = df_donnees_for_dflocal[['Surface reelle bati', 'Nombre pieces principales', 'Valeur fonciere', 'idType', 'idAdresse', 'idMut']].drop_duplicates()
df_local = df_local.reset_index(drop=True)
df_local['idLocal'] = df_local.index.astype(str)+'_L'
df_local
```
